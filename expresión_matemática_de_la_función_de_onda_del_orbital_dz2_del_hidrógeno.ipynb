```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
a0 = 0.529  # Radio de Bohr en Å
Z = 1       # Carga nuclear efectiva para el hidrógeno
r_value = 6 # Valor de r constante
```

```python
# Función de onda radial simplificada para dz^2
def radial_wavefunction(r):
    return 36 * np.exp(-r / a0)

# Valores de r
r_values = np.linspace(0, 10, 100)

# Calcular la función de onda radial para cada valor de r
wavefunction_values = radial_wavefunction(r_values)

# Graficar
plt.plot(r_values, wavefunction_values, label=r'$\Psi_{dz^2}(r)$')
plt.title('Función de onda radial para el orbital dz^2 del hidrógeno')
plt.xlabel('r (Å)')
plt.ylabel(r'$\Psi_{dz^2}(r)$')
plt.axvline(x=r_value, color='r', linestyle='--', label=f'r = {r_value}')
plt.legend()
plt.show()
```
