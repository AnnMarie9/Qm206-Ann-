```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.special import genlaguerre
```

```python
def radial_hydrogen(r, n, l, Z, a0):
    prefactor = np.sqrt((2 * Z / (n * a0))**3 * np.math.factorial(n - l - 1) / (2 * n * np.math.factorial(n + l)**3))
    exponential_term = np.exp(-Z * r / (n * a0))
    laguerre_term = genlaguerre(n - l - 1, 2 * l + 1)((2 * Z * r) / (n * a0))
    radial_function = prefactor * exponential_term * ((2 * Z * r) / (n * a0))**l * laguerre_term
    return radial_function
```

```python
# Parámetros
ao = 1  # Unidad de Bohr
Z = 1   # Número atómico del hidrógeno
r_values = np.linspace(0, 12, 1000)  # Intervalo de r
```

```python
# Graficar la función radial
n = 1  # Número cuántico principal
l = 0  # Número cuántico azimutal
```

```python
radial_values = radial_hydrogen(r_values, n, l, Z, ao)

plt.plot(r_values, radial_values, label=f'n={n}, l={l}')
plt.title('Función Radial del Átomo de Hidrógeno')
plt.xlabel('r (Unidades de Bohr)')
plt.ylabel('$R_{n,l}(r)$')
plt.legend()
plt.show()
```
