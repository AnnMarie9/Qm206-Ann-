# Taller: Cálculo del valor de $ \pi $ mediante la generación de número aleatorios
image.png


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def browniano(x0, y0, n):
    dx = np.random.normal(0, 1, n)
    dy = np.random.normal(0, 1, n)
    dx[0] = x0
    dy[0] = y0
    x2 = np.cumsum(dx)
    y2 = np.cumsum(dy)
    return x2, y2
```

```python
def estimate_pi(num_points):
    inside_circle = 0
    for _ in range(num_points):
        x = np.random.uniform(-20, 20)
        y = np.random.uniform(-20, 20)
        distance = np.sqrt(x**2 + y**2)
        if distance <= 20:  # radio del círculo es 20
            inside_circle += 1
    estimated_pi = (inside_circle / num_points) * 4
    return estimated_pi
```

```python
# Generar puntos aleatorios
x, y = browniano(0, 0, 1000)
```

```python
# Grafico los puntos en el cuadrado
plt.figure(figsize=(8, 8))
plt.scatter(x, y, marker='.', color='blue', alpha=0.5)
plt.title('Puntos Aleatorios en el Cuadrado')
plt.xlabel('X')
plt.ylabel('Y')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
# Valor de π
num_points = 10000
estimated_pi = estimate_pi(num_points)
print(f'Estimación de π usando {num_points} puntos: {estimated_pi}')
```

```python
# gráfico de la estimación de π
plt.figure(figsize=(8, 8))
plt.scatter(x, y, marker='.', color='blue', alpha=0.5)
plt.title(f'Estimación de π: {estimated_pi}')
plt.xlabel('X')
plt.ylabel('Y')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-20, 20)
plt.ylim(-20, 20)
circle = plt.Circle((0, 0), 20, edgecolor='red', facecolor='none', linestyle='dashed')
plt.gca().add_patch(circle)
plt.show()
```
