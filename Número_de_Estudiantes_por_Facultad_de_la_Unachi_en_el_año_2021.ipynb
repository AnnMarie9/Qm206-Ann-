```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
Facultad = ['Arquitectura', 'Enfermería', 'Comunicación Social', 'Economía', 'Administración Pública',
            'Medicina', 'Administración de Empresas y Contabilidad', 'Derecho y Ciencias Políticas',
            'Ciencias Naturales y Exactas', 'Ciencias de la Educación', 'Humanidades']
```

```python
num_estudiantes = [512, 873, 1066, 1476, 1636, 1705, 2852, 2916, 3066, 3665, 4473]
```

```python
palette = sns.color_palette("Set3")
plt.figure(figsize=(10, 6))
sns.barplot(x=num_estudiantes, y=Facultad, palette=palette)
plt.xlabel('Número de Estudiantes')
plt.ylabel('Facultad')
plt.title('UNACHI-Número de Estudiantes por Facultad en El Año 2021')
plt.show()
```
