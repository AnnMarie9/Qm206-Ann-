TALLER: SURFACE PLOTS
Utilizar el c칩digo anterior para generar las gr치ficas correspondientes a las siguientes funciones
*   $z=sin(\sqrt{x^2+y^2}) $
*    $z=(x-1)^2(x+1)^2 + (y-1)^2(y+1)^2$



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Funci칩n z = sin(sqrt(x^2 + y^2))
def z_function(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
# Datos para x, y y z
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x_grid, y_grid = np.meshgrid(x, y)
z = z_function(x_grid, y_grid)
```

```python
# Gr치fica de contorno
contours = plt.contour(x_grid, y_grid, z)
plt.clabel(contours)
plt.title('Contour Plot')
plt.show()
```

```python
# Gr치fica de contorno con nivel de colores y mapa de colores
color_niveles = np.linspace(-1, 1, 20)
contours = plt.contour(x_grid, y_grid, z, levels=color_niveles, colors="black")
plt.clabel(contours)
plt.contourf(x_grid, y_grid, z, levels=color_niveles, cmap='viridis')  # Puedes cambiar 'viridis' por otro mapa de colores
plt.colorbar()
plt.title('Contour Plot with Colored Levels')
plt.show()
```

```python
# Funci칩n z
def func_z(x, y):
    return (x - 1)**2 * (x + 1)**2 + (y - 1)**2 * (y + 1)**2
```

```python
# Puntos
x_grid, y_grid = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
```

```python
# Funci칩n z en los puntos de la malla
z = func_z(x_grid, y_grid)
```

```python
# Gr치ficas de contorno y de contorno relleno
contours = plt.contour(x_grid, y_grid, z, colors="black")
plt.clabel(contours)
```

```python
color_niveles = np.linspace(np.min(z), np.max(z), 20)
contours = plt.contour(x_grid, y_grid, z, levels=color_niveles, colors="black")
plt.clabel(contours)
plt.contourf(x_grid, y_grid, z, levels=color_niveles)
plt.colorbar()
plt.show()
```
